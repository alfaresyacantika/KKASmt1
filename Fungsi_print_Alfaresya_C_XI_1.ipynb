#Parameter Fungsi

```python
nama_siswa_1 = "Favian"
nama_siswa_2 = "Naufal"
nama_siswa_3 = "Humaira"

print(nama_siswa_1, nama_siswa_2, nama_siswa_3, sep= "*")
```

```python
nama_siswa = "Naufal~Favian~Humaira~Favian"
print(nama_siswa, sep= " ")
```

```python
nama_siswa_1 = "Naufal"
nama_siswa_2 = "Favian"
nama_siswa_3 = "Humaira"
nama_siswa_4 = "Favian"

print(nama_siswa_1, end= "~")
print(nama_siswa_2)

print(nama_siswa_3, end= "~")
print(nama_siswa_4)
```
